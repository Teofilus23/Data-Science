# Logistic Regression pada Binary Classification Task 

"Logistic Regression" Dalam statistik, model logistik (atau model logit) digunakan untuk memodelkan probabilitas dari suatu kelas atau event yang ada seperti lulus / gagal, menang / kalah, hidup / mati atau sehat / sakit. Ini dapat diperluas untuk memodelkan beberapa kelas peristiwa seperti menentukan apakah suatu gambar berisi kucing, anjing, singa, dll. Setiap objek yang terdeteksi dalam gambar akan diberi probabilitas antara 0 dan 1, dengan jumlah satu.

Logistic Regression adalah model statistik yang dalam bentuk dasarnya menggunakan fungsi logistik untuk memodelkan variabel dependen biner, meskipun ada banyak ekstensi yang lebih kompleks. Dalam analisis regresi, regresi logistik [1] (atau regresi logit) memperkirakan parameter model logistik (bentuk regresi biner). Secara matematis, model logistik biner memiliki variabel dependen dengan dua kemungkinan nilai, seperti lulus / gagal yang diwakili oleh variabel indikator, dimana kedua nilai tersebut diberi label "0" dan "1". Dalam model logistik, log-odds (logaritma peluang) untuk nilai berlabel "1" adalah kombinasi linier dari satu atau lebih variabel independen ("prediktor"); variabel independen masing-masing dapat berupa variabel biner (dua kelas, dikodekan oleh variabel indikator) atau variabel kontinu (nilai riil apa pun). Probabilitas yang sesuai dari nilai berlabel "1" dapat bervariasi antara 0 (tentu saja nilai "0") dan 1 (tentu saja nilai "1"), karenanya pelabelan; fungsi yang mengubah log-peluang menjadi probabilitas adalah fungsi logistik, karena itulah namanya. Satuan pengukuran untuk skala log-odds disebut logit, dari satuan logistik, maka nama alternatifnya. Model analog dengan fungsi sigmoid berbeda selain fungsi logistik juga dapat digunakan, seperti model probit; karakteristik yang menentukan dari model logistik adalah bahwa meningkatkan salah satu variabel independen secara multiplikasi menskalakan peluang hasil yang diberikan pada tingkat yang konstan, dengan setiap variabel independen memiliki parameternya sendiri; untuk variabel dependen biner, ini menggeneralisasi rasio odds.

# Simple Linear Regression

Dengan "Simple Linear Regression", kita hanya bisa menyertakan satu feature saja untuk melakukan estimasi nilai. Berikut adalah formulanya.

- $y = \alpha + \beta x$
- $g (x) = \alpha + \beta x$

Keterangan :
- $\alpha$ = Merupakan nilai interceptnya.
- $\beta$ = Merupakan nilai slopenya.
- $g (x)$ = $y$ yang merepresentasikan satu buah feature saja.

# Multiple Linear Regression

Pada Multiple Linear Regression, sebuah feature dapat menyertakan lebih dari satu feature untuk melakukan estimasi nilai.

- $y = \alpha + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n$
- $g(X) = \alpha + \beta X$

Berbeda dengan simple linear regression yang merepresentasikan x nya berjumlah tunggal atau satu, jika pada multiple linear regression x nya banyak atau lebih dari satu.

# Logistic Regression

Berikut adalah formula untuk Logistic Regression.

- $g(X) = sigmoid(\alpha + \beta X)$
- $sigmoid(x) = \frac{1}{1 + exp (-x)}$

Model logistic regression tersebut jika kita visualisasikan, maka akan membentuk kurva sigmoid sebagai berikut (perhatikan gambar dibawah ini).

Kurva sigmoid tersebut akan berbentuk seperti huruf S. Rentan nilai dari sumbu x pada ilustrasi disini hanya sekedar contoh saja karena rentan nilainya akan sangat bergantung pada kasus yang sedang dihadapi. Untuk kasus disini rentan nilainya adalah -6 sampai 6. Tetapi nilai rentan ini tidak harus atau mesti -6 sampai 6, rentan nilainya bisa berapapun juga, tergantung pada kasus atau pada dataset yang kita miliki. 

Hal yang menarik dari kurva tersebut adalah pada sumbu y nya. Nilai sumbu y akan selalu berada dama renta nilai 0 dan 1, dan rentan nilai tersebut akan berkolerasi dengan bynomial probability. Pada dasarnya, Logistic Regression ini akan melalukan prediksi nilai yang berada dalam rentan 0 dan 1. Pertanyaan berikutnya adalah, lalu bagaimana dengan kasus binary classification task? sesuai dengan namanya dalam binary classification, hanya terdapat dua buah class saja dan kedua class tersebut bisa kita encode sebagai 0 dan 1. Dengan demikian, ketika nilai probabilitynya mendekati 0, maka akan diklasifikasikan sebagai class 0, sedangkan ketika nilai probabilitynya mendekati 1, maka akan diklasifikasikan sebagai class 1.

# Training & Testing Dataset

Membagi datasetnya kedalam 2 bagian yaitu training dan testing set. Disini pertama-tama lakukan pengelompokkan terlebih dahulu features dan targetnya. Karena targetnya berupa label atau string yang terdiri dari 2 nilai yaitu "ham" dan "spam", maka disini  lakukan pemanfaatan label binarizer untuk melakukan konversi menjadi nilai biner 0 dan 1.

```python
from sklearn.preprocessing import LabelBinarizer

X = df['sms'].values
y = df['label'].values

lb = LabelBinarizer()
y = lb.fit_transform(y).ravel()
lb.classes_
```

- Pertama-tama kita import terlebih dahulu modulnya dengan memanggil method "from sklearn.preprocessing import LabelBinarizer".
- Lalu selanjutnya sekumpulan nilai dari kolom "sms" yang akan ditampung kedalam variabel "X". Lalu sekumpulan nilai dari kolom "label" akan ditampung kedalam variabel "y".
- Selanjutnya bentuk objek dari class label binarizer yang objeknya lalu tampung kedalam variabel "lb".
- Untuk selanjutnya kita akan memanggil "fit_transform" dari objek label binarizer ini yang akan kita kenakan terhadap nilai yang ditampung kedalam variabel "y". Hal lain yang perlu diperhatikan adalah nilai dari variabel "y" setelah dikenakan fit_transform dari label binarizer akan menghasilkan multi dimensional array dan disini kita akan mengkonversikan kembali menjadi array satu dimensi. Oleh karenanya, kita bisa memanggil method ".ravel()" yang hasilnya akan kita tampung ke dalam variabel "y".
- Lalu berikutnya, kita akan mencoba mencari tahu proses label binarizernya. Oleh karenanya, disini kita memangil method "lb.classes_".

Class yang dihandle oleh label binarizernya adalah "ham" dan "spam". Karena "ham" menempati posisi pertama, maka "ham" tersebut akan di encode sebagai 0 sedangkan "spam" akan di encode sebagai 1 atau dengan kata lain, class 0 merupakan class "ham" lalu class 1 merupakan class "spam".

Selanjutnya kita akan pecah dataset ini menjadi 2 buah bagian dengan menggunakan modul train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=0)

print(X_train, '\n')
print(y_train)
```

- Pertama-tama kita mengimport terlebih dahulu modulnya dengan memanggi method "from sklearn.model_selection import train_test_split".
- Selanjutnya kita tinggal memanggil saja train_test_splitnya yang disertai dengan sekumpulan nilai feature dan diikuti dengan sekumpulan nilai labelnya. Disini juga kita spesifikasikan test_sizenya yaitu 0.25 artinya porsi untuk testing datasetnya adalah 25% sedangkan 75% nya akan digunakan sebagai traning set. Lalu kita juga set random_state nya sebagai 0. Pemanggilan train_test_split ini akan menghasilkan 4 kumpulan nilai dan keempat kumpulan nilai tersebut akan kita tampung kedalam variabel X_train, X_test, y_train, dan y_test.
- Lalu terakhir kita akan coba tampilkan X_train dan y_train nya.

training setnya, X_train merupakan nilai feature untuk training setnya sedangkan y_train merupakan nilai label untuk training setnya. Untuk features nya akan berisi data teks sms, sedangkan labelnya akan berisi class yang berisikan nilai 0 dan 1. 


# Feature Extraction dengan TF-IDF

Pada SKLearn, TF-IDF dapat diterapkan dengan manfaatkan modul "TF-IDF Vectorizer".

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print(X_train_tfidf)
```

- Import modulnya terlebih dahulu dengan memanggil method "from sklearn.feature_extraction.text import TfidfVectorizer".
- Lalu selanjutnya kita akan membentuk objek TF-IDF Vectorizer dengan menyertakan parameter "stop_word" yang diberi nilai "english" yang objeknya kita akan tampung kedalam variabel "vectorizer".
- Lalu selanjutnya kita akan memanggi fungsi "fit_transform" dari objek "vectorizer" tersebut yang akan kita kenakan kedalam variabel "X_train" yang hasl transformasinya kita akan tampung kedalam "X_train_tfidf" atau dengan kata lain variabel "X_train_tfidf" tersebut akan menampung sekumpulan nilai "X_train" dan hasil transformasinya kita akan ditampung kedalam variabel "X_train_tfidf" atau dengan kata lain variabel "X_train_tfidf" akan menampung sekumpulan nilai "X_train" yang sudah ditransformasikan menjadi TF-IDF matrix.
- Lalu berikutnya objek vectorizer yang sama juga akan digunakan untuk melakukan proses transformasi terhadap "X_test" dan hasil transformasinya kita akan tampung ke dalam variabel "X_test_tfidf".
- Lalu terakhir kita akan coba tampilkan kelayar "X_train_tfidf" nya.

Nilai yang ditampung bukan lagi berupa teks sms melainkan sudah berupa vector tf-idf.

# Binary Classification dengan Logistic Regression

Menerapkan Logistic Regression untuk melakukan klasifikasi "ham" dan "spam" pada dataset sms yang sudah dimiliki.

- Setelah modelnya ditraining, maka model ini akan menjadi train model dan train model tersebut akan kita gunakan untuk melakuka prediksi dan sekumpulna nilai yang ingin kita prediksi adalah label berdasarkan data-data feature yang ditampung dalam "X_test" TF-IDF yang hasil prediksinya kita akan tampung kedalam variabel "y_pred".

# Terminologi Dasar

Terkait metrix evaluasi pada classification task, terdapat 4 terminologi dasar atau istilah dasar yang perlu dipahami terelebih dahulu.

- True Positive (TP)
- True Negative (TN)
- False Positive (FP)
- False Negative (FN)

Keterangan :
- True Positive dan True Negative, keduanya merepresentasikan hasil prediksi atau hasil klasifikasi yang benar. True Positive berarti sesuatu yang bernilai positif telah dengan tepat diprediksi sebagai positif oleh model, sedangkan  True Negative berarti sesuatu yang bernilai negatif telah dengan tepat diprediksi sebagai negatif oleh model atau jika kita mengacu pada kasus disini, true positif berarti model sudah dengan tepat memprediksi data "spam" sebagai "spam", dan data "ham" sebagai "ham". Sedangkan true negatif berarti model sudah dengan tepat memprediksi data ham sebagai bukan spam dan spam sebagai bukan ham. 
- Dua istilah berikutnya adalah False Positive dan False Negative, keduanya merepresentasikan hasil prediksi atau hasil klasifikasi yang salah. False positive berarti sesuatu yang bernilai negatif telah keliru diprediksi sebagai positif oleh model, sedangkan false negative berarti sesuatu yang bernilai positif telah keliru diprediksi sebagai negatif oleh model atau jika kita mengacu pada kasus kita kali ini, false positive berarti model sudah dengan keliru memprediksi data "ham" sebagai "spam" dan data "spam" sebagai "ham", sedangkan false negative berarti suatu model sudah dengan keliru memprediksi data "spam" sebagai bukan "spam" dan data "ham" sebagai bukan "ham".

# Accuracy

Metrix selanjutnya yang akan kita pelajari adalah metrix "Accuracy" atau akurasi. Accuracy mengukur porsi dari hasil prediksi yang tepat.

$Accuracy = \frac{TP + TN}{TP + TN + FP +FN} = \frac{correct}{total}$

Keterangan :
- TP = True Positive.
- TN = True Negative.
- FP = False Positive.
- FN = False Negative.
- correct = Semua hasil prediksi yang tepat.
- total = Total jumlah prediksi.

Nilai yang dihasilkan oleh matrix accuracy iniakan berada pada rentan nilai 0 dan 1.

Referensi: https://en.wikipedia.org/wiki/Accuracy_and_precision

Selanjutnya kita akan mencoba melakukan pemanfaatan matrix accuracy ini dengan SkLearn.

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

# Precission & Recall

# Precission or Positive Predictive Value (PPV)

Precission juga sering dikenal dengan istilah Positive Predictive Value atau PPV.

$Precission = \frac{TP}{TP + FP}$

Keterangan :
- TP = True Positive.
- FP = False Negative.

Referensi: https://en.wikipedia.org/wiki/Positive_and_negative_predictive_values

Selanjutnya kita akan mencoba memanfaatkan metrix precission dengan SkLearn.

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_pred)
```

- Memanggil method "from sklearn.metrics import precision_score".

# Recall or True Positive Rate (TPR) or Sensitivity

Selanjutnya kita akan mempelajari Recall, Recall sendiri sering dikenal dengan istilah True Positive Rate (TPR). Selain itu, pengukuran Recall juga dikenal sebagai measure of sensitivity. 

$Recall = \frac{TP}{TP + FN}$

Keterangan :
- TP = True Positive.
- FN = True Negative.

Referensi: https://en.wikipedia.org/wiki/Sensitivity_and_specificity

Selanjutnya kita akan melakukan pemanfaatan terhadap metrix recall ini dengan SkLearn.

# F1-Score

Metrix evaluasi selanjutnya yang akan dipelajari adalah F1-score atau F1-measure yaitu harmonic mean dari precission dan recall.

$F1 score = \frac{precission \ x \  recall}{precission \ + \ recall}$

Referensi: https://en.wikipedia.org/wiki/F-score

# ROC: Receiver Operating Characteristic

Metrix evaluasi terakhir yang akan kita pelajari adalah ROC atau Receiver Operating Characteristic. ROC menawarkan visualisasi terhadap performa dari classifier dengan membandingkan nilai Recall (TPR) dan nilai Fallout (FPR).

$fallout = \frac{FP}{TN + FP}$

Keterangan :
- FP = False Positive.
- TN = True Negative.

Referensi: https://en.wikipedia.org/wiki/Receiver_operating_characteristic

Gambar diatas merupakan kurva ROC yang pada dasarnya akan membandingkan 2 buah parameter yaitu True Positive Rate (TPR) dengan False Positive Rate atau (FPR) dimana rentan nilainya sama antara nilai 0 sampai dengan 1. Pada umumnya, kurva ROC akan menampilkan garis linear diagonal semacam tersebut yang garisnya akan merepresentasikan random classifier. Berdasarkan hasil visualisasi dari ROC curve, ketika hasil ploting dari suatu model makin mengarah ke sisi kiri atas, maka peformanya akan dinilai semakin baik, sedangkan ketika ploting modelnya semakin mengarah ke sisi kanan bawah, model tersebut dipandang sebagai model yang buruk. Bisa dilihat ada 3 warna garis biru, orange, dan hijau. Biru hasilnya baik sampai ke hijau hasilnya kurang baik. Suatu model akan mendapat nilai sempurna atau perfect classifier ketika posisinya ada di posisi nya berada di garis merah putus-putus.

Selanjutnya kita akan memanfaatkan ROC ini dengan menggunakan SkLearn.

```python
from sklearn.metrics import roc_curve, auc

prob_estimates = model.predict_proba(X_test_tfidf)

fpr, tpr, threshhold = roc_curve(y_test, prob_estimates[:, 1])
nilai_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, 'b', label=f'AUC={nilai_auc}')
plt.plot([0,1], [0,1], 'r--', label='Random CLassifier')

plt.title('ROC: Receiver Operating Characteristic')
plt.xlabel('Fallout or False Positive Rate')
plt.ylabel('Recall or True Positive Rate')
plt.legend()
plt.show()
```

- Pertama-tama kita akan mengimport terlebih dahulu modulnya dengan memanggil method "from sklearn.metrics import roc_curve, auc". Auc adalah area under the curve yang akan merepresentasikan wilayah atau area yang berada dibawah kurva berwarna biru pada gambar yang akan merepresentasikan proporsi atau presentase area yang berada dibawah curva biru. Bila garis semakin mendekati garis biru atau perfect classifier, maka nilai aoc nya akan semakin mendekati 1.
- Lalu pada selanjutnya kita akan probability estimatenya dengan memanggil "model.predict_proba" dengan menyertakan parameter (X_test_tfidf) yang ditampung kedalam variabel "prob_estimates".
- Lalu selanjutnya untuk roc curve, kita bisa memanggil "roc_curve" dengan menyertakan parameter (y_test, prob_estimates[:, 1]) yang akan mengembalikan 3 buah nilai yaitu "fpr, tpr, threshhold".
- Lalu selanjutnya, berdasarkan nilai fpr dan tpr ini kita bisa menghitung nilai auc nya dengan cara memanggil "auc" dengan menyertakan parameter (fpr, tpr) yang ditampung kedalam variabel "nilai_auc".
- plotting terlebih dahulu garis yang bewarna biru dengan cara memanggil "plt.plot" dengan menyertakan parameter "(fpr, tpr, 'b', label=f'AUC={nilai_auc}')". 'b' untuk membuat warna ploting menajadi biru.
- Lalu selanjutnya kita akan ploting untuk random classifiernya dengan memanggil "plt.plot" dengan menyertakan parameter ([0,1], [0,1], 'r--', label='Random CLassifier'). "r--" berarti garisnya bewarna merah putus-putus.
